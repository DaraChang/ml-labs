```python
import pandas as pd
import numpy as np

#L2Loss function
def L2(yhat, y):
    diff = (y - yhat) ** 2
    loss = np.sum(diff)
    return loss

train_set = pd.read_csv("Lab0/train.csv")
train_set.drop('ID', axis=1, inplace=True)
X = train_set[train_set.loc[:, train_set.columns != 'csMPa'].columns]
Y = train_set['csMPa']
n_iterations=60000
learning_rate=0.0000000001
#éš¨æ©Ÿç”ŸæˆweightçŸ©é™£
w = np.random.randn(8,1)

for iteration in range(n_iterations):
    #æ±‚å‡ºé æ¸¬çš„y_trainå€¼
    predict = np.dot(X,w)
    #èª¤å·®å€¼
    Loss=L2(predict,list(Y))
    Y = np.array(Y).reshape(824, 1)
    output_error =Y-predict
    #gradient descent
    gradients = -2*pd.DataFrame.dot(X.T,output_error)
    w -= learning_rate*gradients
    
X_test = pd.read_csv("Lab0/test.csv")
X_test = X_test[X_test.loc[:, X_test.columns != 'ID'].columns]
#æ±‚å‡ºé æ¸¬çš„y_testå€¼
Y_test = np.dot(X_test,w)
Y_test = pd.DataFrame(Y_test)
Y_test = Y_test.to_csv("0610832_result.csv")
MSE=np.sum(np.power((Y.reshape(-1,1) - predict),2))/len(Y)
print(Loss)
print(MSE)
```
